**TRAINING THE MODEL using FACEBOOK PROPHET**

```python
from fbprophet import Prophet
```

```python
def sales_predictions(Store_ID, sales_df, periods):
    sales_df=sales_df[sales_df['Store']==Store_ID]
    sales_df=sales_df[['Date','Sales']].rename(columns={'Date':'ds','Sales':'y'})
    sales_df=sales_df.sort_values('ds')
    
    model=Prophet()
    model.fit(sales_df)
    future=model.make_future_dataframe(periods=periods)
    forecast=model.predict(future)
    figure=model.plot(forecast,xlabel='Date',ylabel='Sales')
    figure2=model.plot_components(forecast)
    
```

```python
sales_predictions(10,sales_all_df,60)
```

```python
 # Function takes in the data frame, storeID, and number of future period forecast
 # The function then generates date/sales columns in Prophet format
 # The function then makes time series predictions
```

**Predicting Future With FBPROPHET  considering holidays as well**1


```python
def sales_predictions(Store_ID, sales_df, holidays, periods):
    sales_df=sales_df[sales_df['Store']==Store_ID]
    sales_df=sales_df[['Date','Sales']].rename(columns={'Date':'ds','Sales':'y'})
    sales_df=sales_df.sort_values('ds')
    
    model=Prophet(holidays=holidays)
    model.fit(sales_df)
    future=model.make_future_dataframe(periods=periods)
    forecast=model.predict(future)
    figure=model.plot(forecast,xlabel='Date',ylabel='Sales')
    figure2=model.plot_components(forecast)
    
```

```python
# Getting all the dates pertaining to school holidays 

school_holidays=sales_all_df[sales_all_df['SchoolHoliday']==1].loc[:,'Date'].values
```

```python
# Getting all the dates pertaining to state holidays 

state_holidays=sales_all_df[(sales_all_df['StateHoliday']=='a')|(sales_all_df['StateHoliday']=='b')|(sales_all_df['StateHoliday']=='c')].loc[:,'Date'].values
```

```python
state_holidays
```

```python
state_holidays=pd.DataFrame({'ds':pd.to_datetime(state_holidays),'holiday':'state_holiday'})

```

```python
state_holidays
```

```python
school_holidays = pd.DataFrame({'ds': pd.to_datetime(school_holidays),
                                'holiday': 'school_holiday'})
```

```python
school_holidays
```

```python
#concatenate both school and state holidays
school_state_holidays=pd.concat((state_holidays,school_holidays))
```

```python
school_state_holidays
```

```python
# Making predictions using holidays for a specific store

sales_predictions(6,sales_all_df,school_state_holidays,90)
```
